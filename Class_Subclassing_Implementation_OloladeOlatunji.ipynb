```python
# First, import the library we need to plot graphs
import matplotlib.pyplot as plt

# Create a general class called Machine
class Machine:
    def __init__(self, name):
        self.name = name  # The machine has a name

    def describe(self):
        print(f"This is a machine called {self.name}")

# Now, create a subclass for Pump
class Pump(Machine):
    def __init__(self, name, flow_rate):
        super().__init__(name)  # Inherit name from Machine
        self.flow_rate = flow_rate  # Pump has a flow rate

    def pressure_increase(self, time):
        # Simple rule: Pressure = Flow rate * Time
        return self.flow_rate * time

# Create another subclass for Compressor
class Compressor(Machine):
    def __init__(self, name, compression_ratio):
        super().__init__(name)  # Inherit name from Machine
        self.compression_ratio = compression_ratio  # Compressor has a compression ratio

    def pressure_increase(self, time):
        # Different rule: Pressure grows a bit faster
        return (self.compression_ratio * time) ** 1.5

# Now, let's create objects
pump1 = Pump("Water Pump", flow_rate=2)
compressor1 = Compressor("Air Compressor", compression_ratio=1.2)

# Let's prepare time values
time_values = [i for i in range(11)]  # time from 0 to 10

# Calculate pressure for each machine
pump_pressures = [pump1.pressure_increase(t) for t in time_values]
compressor_pressures = [compressor1.pressure_increase(t) for t in time_values]

# Now, let's plot the pressures
plt.plot(time_values, pump_pressures, label="Pump Pressure")
plt.plot(time_values, compressor_pressures, label="Compressor Pressure")

# Add labels and title
plt.xlabel("Time (seconds)")
plt.ylabel("Pressure")
plt.title("Pressure Increase Over Time")
plt.legend()

# Show the graph
plt.show()
```
