```python
# pip install matplotlib

import matplotlib.pyplot as plt  # This helps us to draw charts
```

```python
# Line chart to show temperature changing over time

# Time in hours
time = [1, 2, 3, 4, 5]

# Temperature readings
temperature = [30, 32, 35, 33, 31]

# Plotting the line
plt.plot(time, temperature)

# Giving the chart a title
plt.title("Temperature Over Time")

# Label for the x-axis
plt.xlabel("Time (hours)")

# Label for the y-axis
plt.ylabel("Temperature (°C)")

# Show the chart
plt.show()
```

```python
# Bar chart to compare the number of boys and girls

# Categories
categories = ["Boys", "Girls"]

# Number of boys and girls
values = [12, 18]

# Drawing the bars
plt.bar(categories, values)

# Giving the chart a title
plt.title("Number of Boys and Girls")

# Label for the y-axis
plt.ylabel("Count")

# Show the chart
plt.show()
```

```python
# Pie chart to show how we spend our day

# Different activities
activities = ["Sleep", "School", "Fun", "Homework"]

# Hours spent on each activity
hours = [8, 6, 5, 5]

# Drawing the pie chart
plt.pie(hours, labels=activities, autopct='%1.1f%%')

# Title for the chart
plt.title("Daily Activities")

# Show the chart
plt.show()
```

```python
# Scatter plot to show relationship between height and weight

# Heights of students
height = [150, 160, 165, 155, 170]

# Weights of students
weight = [45, 50, 55, 48, 60]

# Drawing the dots
plt.scatter(height, weight)

# Title of the chart
plt.title("Height vs Weight")

# Labels for the axes
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")

# Show the chart
plt.show()
```

```python
# Histogram to show how student scores are distributed

# List of scores
scores = [45, 55, 60, 65, 70, 50, 48, 75, 80, 90]

# Drawing the histogram
plt.hist(scores, bins=5)

# Title of the chart
plt.title("Score Distribution")

# Labels for the axes
plt.xlabel("Score Range")
plt.ylabel("Number of Students")

# Show the chart
plt.show()
```
