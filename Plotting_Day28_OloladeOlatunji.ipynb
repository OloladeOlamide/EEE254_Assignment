```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))

# Generate numbers and their squares
numbers = np.arange(1, 1001) #this genrates the numbers 1-1000
squares = numbers**2  #this generates the square of the numbers

# Plotting
plt.plot(numbers, squares, color='purple')
plt.title('Squares of Numbers from 1 to 1000')
plt.xlabel('Number')
plt.ylabel('Square of Number')
plt.grid(True)
plt.show()
```
